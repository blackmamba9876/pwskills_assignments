```python
# required packages 

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# data ingestion.
df=pd.read_csv('dataset.csv')
```

```python
df=df.drop_duplicates()
```

UniVariate Analysis

```python
# Bar plot for Gender distribution
sns.countplot(x='GENDER', data=df, palette='coolwarm')
plt.title('Gender Distribution')
plt.show()
```

Observation: This plot shows the distribution of gender in the dataset, indicating the proportion of male (M) and female (F) participants.

```python
# Histogram for Age distribution
plt.hist(df['AGE'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Observation: This histogram reveals the age distribution among the participants, highlighting the age groups that are most represented in the dataset.

```python
# Pie chart for Lung Cancer cases
df['LUNG_CANCER'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Lung Cancer Cases')
plt.ylabel('')
plt.show()
```

Observation: The pie chart displays the proportion of participants diagnosed with lung cancer versus those who are not.

```python
# Bar plot for Smoking habit
sns.countplot(x='SMOKING', data=df, palette='Set2')
plt.title('Smoking Habit Distribution')
plt.show()
```

Observation: The bar chart shows the distribution of smoking habits among participants, highlighting how many are smokers (1) and non-smokers (2).

Bivariate Analysis

```python
# Box plot for Age vs Lung Cancer
sns.boxplot(x='LUNG_CANCER', y='AGE', data=df, palette='Set3')
plt.title('Age Distribution vs Lung Cancer')
plt.show()
```

Observation: This box plot compares the age distribution between participants diagnosed with lung cancer and those without it, helping to identify any age-related trends.

```python
# Bar plot for Smoking vs Lung Cancer
sns.countplot(x='SMOKING', hue='LUNG_CANCER', data=df, palette='coolwarm')
plt.title('Smoking Habit vs Lung Cancer')
plt.show()
```

Observation: This bar plot highlights the relationship between smoking habits and lung cancer occurrence, indicating whether smoking is a significant factor.

```python
# Bar plot for Gender vs Lung Cancer
sns.countplot(x='GENDER', hue='LUNG_CANCER', data=df, palette='Set1')
plt.title('Gender vs Lung Cancer')
plt.show()
```

Observation: This plot shows the relationship between gender and lung cancer, indicating any differences in lung cancer prevalence between males and females.

 Multivariate Analysis

```python
# Pairplot for key variables
sns.pairplot(df, hue='LUNG_CANCER', vars=['AGE', 'SMOKING', 'YELLOW_FINGERS'], palette='coolwarm')
plt.show()
```

Observation: The pairplot visually explores relationships between age, smoking, and yellow fingers in the context of lung cancer, revealing potential interactions among these variables.

```python
# Heatmap for correlation between variables
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

Observation: The heatmap highlights the correlations between various features, showing which variables are strongly associated with each other, which can provide insights into factors contributing to lung cancer.
